```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

def plot_data(dilations, mape_lr_001, mape_lr_0005, mape_lr_0001, figure_name):
    # 设置全局字体和字体大小
    plt.rcParams['font.family'] = 'Times New Roman'
    plt.rcParams['axes.labelsize'] = 20  # x和y标签的字体大小
    plt.rcParams['xtick.labelsize'] = 18  # x轴刻度的字体大小
    plt.rcParams['ytick.labelsize'] = 18  # y轴刻度的字体大小
    plt.rcParams['legend.fontsize'] = 18  # 图例的字体大小

    # 创建图像和轴对象
    fig, ax = plt.subplots(figsize=(5, 4))  # 图片大小可以根据实际需要调整

    # 绘制折线图
    ax.plot(dilations, mape_lr_001, marker='o', label='lr=0.01')
    ax.plot(dilations, mape_lr_0005, marker='o', label='lr=0.005')
    ax.plot(dilations, mape_lr_0001, marker='o', label='lr=0.001')

    # 设置X轴的刻度位置和刻度标签
    ax.xaxis.set_major_locator(ticker.FixedLocator(dilations))
    ax.xaxis.set_major_formatter(ticker.FixedFormatter(dilations))
    
    # 设置x轴刻度全部显示，y轴刻度显示至少5个
    # ax.xaxis.set_major_locator(ticker.MaxNLocator(integer=True))
    ax.yaxis.set_major_locator(ticker.MaxNLocator(nbins=7))
    
    # 添加标题和标签
    ax.set_xlabel('dilation')
    ax.set_ylabel('MAPE (%)')

    # 显示图例
    ax.legend(loc='upper right', frameon=True)  # 有边框的图例
    plt.grid(True)
    # 调整边距以避免边缘的剪切
    plt.tight_layout()

    plt.savefig(figure_name, dpi=300)
```
